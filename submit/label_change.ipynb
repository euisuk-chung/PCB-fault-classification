```python
import os
import pandas as pd
import numpy as np
```

```python
path = os.getcwd()
```

```python
file_path = os.path.join(path, 'submission_model_7_prob.csv')
#file_path = '/home/sks/COMPETITION/DACON/computer_vision2/dataset/dirty_mnist_2nd_answer.csv'
```

```python
probs = pd.read_csv(file_path)
probs.head(3)
```

```python
counts = []
for i in range(len(probs)):
    labels = probs.iloc[i, 1:].values
    label_sum = sum(labels)
    counts.append(label_sum)
```

```python
fin_labels = []

for i in range(len(probs)):
    
    A = probs.iloc[i, 1:].values
    
    sum_labels = sum(A >= 0.5)
    if sum_labels >= 10 and sum_labels <= 14:
        A = np.where(A >= 0.5, 1, 0)
        fin_labels.append(A)
    elif sum_labels < 10:
        one_index = np.argsort(-A)[:10]
        zero_index = np.argsort(-A)[10:]
        A[one_index] = 1
        A[zero_index] = 0
        fin_labels.append(A)
    else:
        one_index = np.argsort(-A)[:14]
        zero_index = np.argsort(-A)[14:]
        A[one_index] = 1
        A[zero_index] = 0
        fin_labels.append(A)
```

```python
fin_probs = np.array(fin_labels)
```

```python
probs.iloc[:, 1:] = fin_probs
```

```python
probs.iloc[:, 1:] = probs.iloc[:, 1:].astype(int)
```
